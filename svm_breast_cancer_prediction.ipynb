```python
X= df.drop('diagnosis',axis=1).values #bağımsız değişken 
y = df['diagnosis'].values #bağımlı değişken
```

```python
#sınıf etiketlerinin tam sayılara (0,1) dönüştürülmesi
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y= le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split #veri setinin eğitim-test olark bölünmesi
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler #0-1 dönüşümü
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# Svm(support vektör makinesi)

```python
from sklearn.svm import SVC
svc_model = SVC(gamma='auto',kernel="linear") #kernel'idoğrusal yani lineer olarak ayarladık
svc_model.fit(X_train, y_train)
```

```python
# Eğitim modelini tahmin etme 
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, y_predict)
```

```python
cm
```

```python
#heatmap (gerçek değerler ve tahmin edilen değerler ısı haritası)
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_predict) #roc auc skorunu gerçek değerler ve tahmin edilen değerler için bulundu
```

```python
accuracy_score(y_test, y_predict) #başarı oranı
```

```python
from sklearn.model_selection import cross_val_score  #10 katlı cross validation
cross_val_score(svc_model, X_test, y_test, cv=10).mean() #10 katlı cross validation ortalaması
```

# Model Tuning

```python
svm=SVC()
```

```python
svm_params={"C":np.arange(1,10), "kernel":["linear","rbf"]} #hiperparametre değerleri
#hangi kernel değeri ve hangi paramtre ddeğeri bu model için uygun yukarıda verdiğimiz alternatiflerden?
```

```python
from sklearn.model_selection import GridSearchCV #GridSearchCV dahil edildi
svm_cv_model=GridSearchCV(svm,svm_params, cv=10,n_jobs=-1, verbose=2).fit(X_train,y_train) 
#10  katlı cross validation yapılacak ve en ugun parametre değeri belirlemesi GridSearchCV ile elde edilecek
```

```python
#180 fits(180 tane fit etme işlemi gerçeklştirilecek bu bilgiyi verbose argümanı ile elde ettik)
```

```python
svm_cv_model.best_score_ #en iyi skore değeri 
#modelin en iyi skor değeri
```

```python
svm_cv_model.best_params_ #svm cv modelin en iyi parametre değerleri
#best parametre değeri sonucu en iyi ceza parametresi (C) olarak 2 değeri ile ve kernel'ıda rbf olarak seçti
```

# final model

```python
#oluşturduğumuz final modeli üzerinden test hatasına erişeceğiz
svm_tuned=SVC(C=2, kernel="rbf").fit(X_train,y_train) #tuned edilmiş model
```

```python
y_pred=svm_tuned.predict(X_test) 
```

```python
accuracy_score(y_test,y_pred)
```
